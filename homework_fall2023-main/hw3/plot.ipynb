```python
import matplotlib.pyplot as plt
import json

# Function to load data from JSON files
def load_data(file_path):
    with open(file_path, 'r') as file:
        return json.load(file)

# Function to average performance over a specified number of points
def average_performance(data, num_points):
    return [sum(data[max(i-num_points+1, 0):i+1]) / min(i+1, num_points) for i in range(len(data))]

# Paths to the Pacman JSON files
json_files = {
    "Pacman_lr0.0001_TrainReturn": "/home/jimchen/Downloads/data_hw3_dqn_dqn_MsPacmanNoFrameskip-v0_d0.99_tu2000_lr0.0001_doubleq_clip10.0_25-11-2023_06-49-01 (1).json",
    "Pacman_lr0.0001_EvalReturn": "/home/jimchen/Downloads/data_hw3_dqn_dqn_MsPacmanNoFrameskip-v0_d0.99_tu2000_lr0.0001_doubleq_clip10.0_25-11-2023_06-49-01.json",
}

# Load data from files
data = {label: load_data(path) for label, path in json_files.items()}

plt.figure(figsize=(12, 8))

# Specify the number of points for averaging
num_points_for_averaging = 10

for label, points in data.items():
    iterations = [point[1] for point in points]
    performance = [point[2] for point in points]
    if "TrainReturn" in label:
        # Averaging performance for train return
        performance = average_performance(performance, num_points_for_averaging)

    plt.plot(iterations, performance, label=label)

plt.xlabel(" Environment Steps")
plt.ylabel("Performance Measure")
plt.title(f"MsPacman TrainReturn vs EvalReturn")
plt.legend()
plt.grid(True)
plt.show()
```
