```python
# Modified script to average the data from the five Inverted Pendulum JSON files

import matplotlib.pyplot as plt
import json
import numpy as np

# Function to load data from JSON files
def load_data(file_path):
    with open(file_path, 'r') as file:
        return json.load(file)

# Paths to the specified Inverted Pendulum JSON files
json_files = [
    "/home/jimchen/Downloads/data_q2_pg_pendulum_default_s1_InvertedPendulum-v4_21-11-2023_00-56-19.json",
    "/home/jimchen/Downloads/data_q2_pg_pendulum_default_s2_InvertedPendulum-v4_21-11-2023_00-58-02.json",
    "/home/jimchen/Downloads/data_q2_pg_pendulum_default_s3_InvertedPendulum-v4_21-11-2023_00-59-44.json",
    "/home/jimchen/Downloads/data_q2_pg_pendulum_default_s4_InvertedPendulum-v4_21-11-2023_01-01-27.json",
    "/home/jimchen/Downloads/data_q2_pg_pendulum_default_s5_InvertedPendulum-v4_21-11-2023_01-03-09.json"
]

# Load data from files and prepare for averaging
all_data = [load_data(path) for path in json_files]

# Assuming that each iteration corresponds to a certain number of environment steps
env_steps_per_iter = 1000  # Adjust this based on your experiment setup
max_env_steps = 2e+5       # Adjust this based on your experiment setup
max_iters = int(max_env_steps / env_steps_per_iter)

# Initialize lists to store averaged data
average_iterations = []
average_returns = []

# Iterate over each iteration
for i in range(max_iters):
    # Collect data for each file at iteration i
    iter_data = [file_data[i] for file_data in all_data if i < len(file_data)]
    
    # Calculate average iteration and return values
    avg_iteration = np.mean([data[1] for data in iter_data]) * env_steps_per_iter
    avg_return = np.mean([data[2] for data in iter_data])

    average_iterations.append(avg_iteration)
    average_returns.append(avg_return)

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(average_iterations, average_returns, label="Average Performance")

plt.xlabel("Environment Steps")
plt.ylabel("Average Performance Metric")
plt.title("Average Performance (5 seeds average) for Inverted Pendulum")
plt.legend()
plt.grid(True)
plt.xlim(0, max_env_steps)
plt.show()
```
