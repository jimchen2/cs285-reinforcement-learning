```python
import matplotlib.pyplot as plt
import json

# Function to load data from JSON files
def load_data(file_path):
    with open(file_path, 'r') as file:
        return json.load(file)

# Path to the CartPole JSON file
json_file = {
    "Awac_Easy": "/home/jimchen/Downloads/hw5_offline_PointmassEasy-v0_awac10.0_28-11-2023_20-28-38_..json",
    "Awac_Medium": "/home/jimchen/Downloads/hw5_offline_PointmassMedium-v0_awac10.0_28-11-2023_20-38-45_..json",
}

# Load data from file
data = {label: load_data(path) for label, path in json_file.items()}

plt.figure(figsize=(10, 6))

for label, points in data.items():
    # Extracting iterations (or environment steps) and performance measure from each point
    iterations = [point[1] for point in points]  # Assuming point[1] represents iterations or steps
    performance = [point[2] for point in points]  # Assuming point[2] represents performance measure

    plt.plot(iterations, performance, label=label)

plt.xlabel(" Environment Steps")
plt.ylabel("Performance Measure")
plt.title("Performance vs  Environment Steps for Awac")
plt.legend()
plt.grid(True)
plt.show()
```
